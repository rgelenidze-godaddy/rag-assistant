```python
from brain.embedding import instance

# Get singleton embedder
embedder = instance.get_instance()
```

```python
s1 = "Alice teaches a course programming 101"
s2 = "Alice is a lecturer of the introduction to programming course"

# Embed the texts
vec1 = embedder.embed_query(s1)
vec2 = embedder.embed_query(s2)

print(vec1, vec2)
```

```python
# Compute cosine similarity (1 - cosine distance)
from scipy.spatial.distance import cosine  # use cosine distance from scipy

similarity = 1 - cosine(vec1, vec2)

print(f"Similarity: {similarity:.4f}")
```
